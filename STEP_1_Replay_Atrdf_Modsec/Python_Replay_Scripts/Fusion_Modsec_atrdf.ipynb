```python
import json
import os

# Chemins
atrdf_path = "/content/drive/MyDrive/BINOME_WORK/PFE_NIT/AI_CODE_DATASETS/STEP_1_Replay_Atrdf_Modsec/Datasets/atrdf.json"
modsec_path = "/content/drive/MyDrive/BINOME_WORK/PFE_NIT/AI_CODE_DATASETS/STEP_1_Replay_Atrdf_Modsec/Datasets/modsec_audit.json"
output_path = "/content/drive/MyDrive/BINOME_WORK/PFE_NIT/AI_CODE_DATASETS/STEP_1_Replay_Atrdf_Modsec/Datasets/modsec_atrdf_joined.json"

# Charger le dataset d'origine
with open(atrdf_path, "r", encoding="utf-8") as f:
    atrdf = json.load(f)

# Charger le log parsé
with open(modsec_path, "r", encoding="utf-8") as f:
    logs = json.load(f)

# Index du dataset original par primary_key
pk_to_entry = {entry["primary_key"]: entry for entry in atrdf}

# Ajout de la requête d'origine à chaque log
for log in logs:
    pk = log.get("primary_key")
    log["original_request"] = pk_to_entry.get(pk)

# Créer le fichier joint s'il n'existe pas, sinon l'écraser (comportement par défaut)
with open(output_path, "w", encoding="utf-8") as f:
    json.dump(logs, f, indent=2, ensure_ascii=False)

if os.path.exists(output_path):
    print(f"[INFO] Fichier fusionné créé ou écrasé : {output_path}")
else:
    print(f"[ERREUR] Impossible de créer le fichier fusionné !")
```
